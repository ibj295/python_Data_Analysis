```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Human Resources.csv")
```

```python
df.shape
```

```python
Emp_count = df.groupby('location_state')['id'].count().reset_index()
Emp_count.columns = ['state','Emp_count']
print(Emp_count)
```

```python
Emp_count.plot(x = 'state', y = 'Emp_count', kind = 'bar', legend=False)
plt.xlable = ('state')
plt.ylable = ('Emp_count')
plt.title = ('Employee Count by state')
plt.show()
```

```python
Total_Emp_Count = df['id'].count()
print(f"Total Employee count is: {Total_Emp_Count}")
```

```python
Gender_count = df['gender'].value_counts()
print(Gender_count)
```

```python
Gender_count.plot(x = 'state', y = 'Emp_count', kind = 'bar', legend=False)
plt.xlable = ('state')
plt.ylable = ('Emp_count')
plt.title = ('Employee Count by state')
plt.show()
```

```python
# Convert 'Hire_Date' to datetime format
df['hire_date'] = pd.to_datetime(df['hire_date'])

# Extract year from 'Hire_Date'
df['Hire_Year'] = df['hire_date'].dt.year

# Count employees hired in each year
yearly_hiring_count = df['Hire_Year'].value_counts().sort_index()

yearly_hiring_count.columns = ['Year', 'Count']


print(yearly_hiring_count)
```

```python
yearly_hiring_count.plot(kind = 'bar')
```

```python
Race_analysis = df['race'].value_counts()
print(Race_analysis)
```

```python
Race_analysis.plot( y = 'values', kind = 'pie' , labels=df['race'], autopct='%1.1f%%', legend=False, title = 'Race % by Race')
```
